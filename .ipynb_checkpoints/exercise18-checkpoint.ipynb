```python
import pandas as pd
import numpy as np
```

```python
data_frame = {}

file_names = ['AAPL', 'GOOG', 'META', 'MSFT']

for file_name in file_names:
    file_path = f'data_new/{file_name}.csv'
    data_frame[file_name] = pd.read_csv(file_path)

aapl_data = data_frame['AAPL'][['Date', 'Close']].rename(columns = {'Close': 'AAPL'})
goog_data = data_frame['GOOG'][['Date', 'Close']].rename(columns = {'Close': 'GOOG'})
meta_data = data_frame['META'][['Date', 'Close']].rename(columns = {'Close': 'META'})
msft_data = data_frame['MSFT'][['Date', 'Close']].rename(columns = {'Close': 'MSFT'})
```

```python
merged_data = pd.merge(aapl_data, goog_data, 
                          on = "Date", 
                          how = 'outer'
                         ).merge(meta_data, 
                          on = "Date",
                          how = 'outer'
                         ).merge(msft_data,
                          on = 'Date',
                          how = 'outer'
                         )
merged_data
```

```python
merged_data.to_csv('data_new/stocks.csv', sep = ",", decimal = ".")
```
